```python
from huawei_graph_tool import my_random_circuit
from Test_tool import Test_ansatz
import random
import numpy as np
from mindquantum.algorithm.nisq import HardwareEfficientAnsatz  
from mindquantum.core.gates import H, RX, RY, RZ,X     
from EHA_Ansatz import EHA
```

```python
import numpy as np

cir_history = []
acc_history = []

for i in range(100):
    rng = np.random.default_rng()  # 创建新的随机数生成器
    random_gate_num = 40
    random_seed = rng.integers(0, 6000)
    sd_rate = rng.uniform(0.3, 0.8)
    ctrl_rate = rng.uniform(0.1, 0.2)
    
    # 打印随机数以检查是否重复
    print(f'Random Seed: {random_seed}, SD Rate: {sd_rate}, CTRL Rate: {ctrl_rate}')
    
    print(f'ansatz_info:seed={random_seed}')
    cir = my_random_circuit(n_qubits=8, gate_num=random_gate_num, sd_rate=sd_rate, ctrl_rate=ctrl_rate, seed=random_seed)
    acc = Test_ansatz(ansatz=cir)
    acc_history.append(np.max(acc[-5:]))
    cir_history.append(cir)
    print(f'第{i}个ACC={np.max(acc[-5:])}')

```

```python
np.max(np.array(acc_history))
```
