```python
import networkx as nx
import matplotlib.pyplot as plt

# 创建一个空图
G = nx.Graph()

# 添加节点
G.add_nodes_from([1, 2, 3, 4, 5, 6])

# 添加边
G.add_edges_from([(1, 2), (1, 3), (2, 4), (2, 5), (3, 6)])

# 手动指定每个节点的位置
pos = {
    1: (0, 2),  # 第一层
    2: (-1, 1),  # 第二层
    3: (1, 1),  # 第二层
    4: (-1.5, 0),  # 第三层
    5: (-0.5, 0),  # 第三层
    6: (1, 0)  # 第三层
}

# 绘制图
nx.draw(G, pos, with_labels=True, node_size=700, node_color='lightblue', font_size=10, font_color='black')
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# 创建一个空图
G = nx.Graph()

# 添加节点
G.add_nodes_from([1, 2, 3])

# 添加边
G.add_edges_from([(1, 2), (1, 3)])

# 手动指定每个节点的位置
pos = {
    1: (-3, 0),  # 第一层
    2: (-2, 1),  # 第二层
    3: (-2, -1),  # 第二层
}

# 绘制图
nx.draw(G, pos, with_labels=True, node_size=700, node_color='lightblue', font_size=10, font_color='black')
plt.show()

adj_matrix = nx.adjacency_matrix(G).todense()
print("邻接矩阵:")
print(adj_matrix)
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# 创建一个有向图对象
G = nx.DiGraph()

# 添加节点和边
G.add_edges_from([(1, 2), (1, 3), (2, 4), (3, 4), (4, 5)])

# 计算节点的位置
pos = nx.spring_layout(G)

# 绘制有向无环图
nx.draw(G, pos, with_labels=True, node_size=700, node_color='lightblue', font_size=10, font_color='black', arrows=True)
plt.show()
```
