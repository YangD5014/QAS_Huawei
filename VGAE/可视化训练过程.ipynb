```python
import matplotlib.pyplot as plt
import numpy as np
import json
```

```python
with open('./result_1015.json','r') as f:
    data = json.load(f)
```

```python
plt.plot(range(len(data['Loss'])),data['Loss'],label='Loss')
# plt.plot(range(len(data['AP'])),data['AP'],label='AP')
plt.legend()
plt.title('VGAE Training ')
plt.xlabel('Epoch')
plt.ylabel('Loss')
```

```python
plt.plot(range(len(data['AP'])),data['AP'],label='AP')
plt.plot(range(len(data["ROC"])),data['ROC'],label='ROC')
# plt.plot(range(len(data['AP'])),data['AP'],label='AP')
plt.legend()
plt.title('VGAE Training ')
plt.xlabel('Epoch')
```

```python
fig, axs = plt.subplots(2, 1, figsize=(6, 8))

# Loss
axs[0].plot(range(len(data['Loss'])), data['Loss'], label='Loss')
axs[0].legend()
axs[0].set_title('VGAE Training  (22|32|8)')
axs[0].set_xlabel('Epoch')
axs[0].set_ylabel('Loss')

# AP  ROC
axs[1].plot(range(len(data['AP'])), data['AP'], label='AP')
axs[1].plot(range(len(data['ROC'])), data['ROC'], label='AUC')
axs[1].legend()
axs[1].set_title('VGAE Training - AP and AUC')
axs[1].set_xlabel('Epoch')
axs[1].set_ylabel('Score')

# 
plt.tight_layout()
plt.show()
```
